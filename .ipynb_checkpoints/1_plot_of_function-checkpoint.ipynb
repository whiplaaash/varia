## Section 1: Plot of a function

```python
import numpy as np
import matplotlib.pyplot as plt
import math as math
import seaborn as sns
sns.set()

# Step 1: Define the function you want to plot
def f(x):
    return np.exp(1/x)/((x**2)*(np.exp(1/x)-1)**2) # example function

# Step 2: Generate x values
x = np.linspace(0.1, 20, 100)  # Generates points from 0 to 400

# Step 3: Compute y values using the function
y = f(x)

#g = np.exp(1/x)
#h = 1/(x**2)

# Step 4: Create the plot
plt.plot(x, y, label="y", lw=3, c="red")

# Step 5: Add labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0,20)
plt.ylim(0,1.2)
plt.title("Plot of y")
plt.legend()

# Step 6: Show the plot
plt.show()
```
